Faça um programa em python pra ler o nome de duas notas de uma turma com "x" numeros de alunos. 
O programa terá fim quando o nome do aluno for "FIM".
Para cada aluno, deverá ser calculada a média.
A média deverá ser calculada por meio de uma função propria pra este fim, chamada Media(). Essa função deverá retornar o valor da média calculada.
Para cada aluno, deverá ser informado a sua situação, a saber:
 Para média menor que 6: "Aluno em Recuperação"
 Para média entre 6 e 7 (exclusive): "Aluno Regular"
 Para média entre 7 (inclusive) e 9 (exclusive): "Aluno Bom"
 Para aluno com média maior ou igual a 9: "Aluno Excelente"
A situação do aluno deverá ser execuada por meio de uma função propria, chamada Situacao(). Essa função dverá retornar a mensagem.
Ao final do programa, deverá ser informada a média da turma. Essa informação deverá ser feita dentro do proprio programa, nao precisando ser invocada qualquer função pra este fim.
