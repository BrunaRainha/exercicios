```python
#estruturas condicionais (lembrar de operadores condicionais e conectivos logicos - slide/foto cm explicação)

# < 16 nao pode votar
# >= 16 E < 18 opcional
# >= 18 E < 65 obrigatorio
# >= 65 opcional

idade = int(input('informe a sua idade:'))

if idade < 16:
  print('nao pode votar')
elif idade >= 16 and idade < 18 or idade >= 65:   #elif é else if
  print('voto opcional')
else:
  print('voto obrigatorio')
```

```python
#repeticao condicional while (numeros pares entre 0 e 100)

numero = 0
while(numero < 100):
  if(numero % 2 == 0):
    print(numero)
  numero = numero + 1
```

```python
#repeticao for (numeros pares entre 0 e 100)

for x in range(0,100):
  if (x % 2 == 0):
    print(x)
```

```python
#repeticao for (numeros pares entre 0 e 100)

for x in range(0,100):
  if(x % 2 != 0):
    print(x)
```

```python
#listas

nomes = ['ana', 'paulo', 'fernanda', 'joao', 57]  #a lista é dinamica
#print(nomes[2])   #especifico, contagem começa do 0

for nome in nomes:  #toda a lista
  print(nome)
```

```python
print('fernanda' in nomes)
print('bruna' in nomes)
print('57' in nomes)
```

```python
#funções em listas

nomes.append('mariana') #adiciona no final da lista

nomes.insert(4, 'clara') #adiciona em uma determinada posicao

nomes.pop() #remove o ultimo elemento da lista

nomes.remove('clara') #remove um elemento da lista

#entre outras funções

#depois colocar
for nome in nomes:  #toda a lista
  print((nome))
```
