```python
#programa pra ler o nome de um aluno

nome_aluno=input("informe o nome do aluno: ")
print(nome_aluno)

#ler mais a nota do aluno
nota1=input("primeira nota do aluno: ")
nota2=input("segunda nota do aluno: ")

#eval = transforma letra em numero
nota1=eval(nota1)
nota2=eval(nota2)

#ou apenas
#nota1=int(nota1)
#nota2=int(nota2)

media=(nota1+nota2)/2
print(f"a média do candango: {nome_aluno} é igual a {media}")

#com media menor que 6 é reprovado, se for maior é aprovado

if(media<6):
  print("reprovado")
else:
  print("aprovado")

```
