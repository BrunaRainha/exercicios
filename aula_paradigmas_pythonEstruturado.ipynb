```python
continua = input('Deseja continuar(Sim/Não)?') #estrutura de repetição
while continua.upper() == 'SIM':   #.upper nao importa o que eu escrever, ele vai colocar em maiusculo
  continua = input('Deseja continuar(Sim/Não)?')

print(f'voce digitou {continua} e agora está fora do laço')
```

```python
while True:  #estrutura de repetição de outra forma (resposta mais lenta)
  continua = input('Deseja continuar(Sim/Não)?')
  if continua.upper() == "NÃO":
    break

print(f'voce digitou {continua} e agora está fora do laço')
```
