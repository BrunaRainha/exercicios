```python
#20 pessoas votam entre a ou b ao final deve dizer quantos votos cada um levou e quem ganhou

lista = []
totalA = 0
totalB = 0

for x in range(5):
 voto = input('escolha entre os candidatos a ou b:')
 lista.append(voto)
print(lista)

#lower - converte uma string pra minuscula
#upper - converte uma string pra maiuscula

for x in lista:
  if (str.upper(x) == 'A'):
    totalA += 1
  elif (str.upper(x) == 'B'):
    totalB += 1

if (totalA > totalB):
  print('canditado a venceu')
elif (totalB > totalA):
  print('canditado b venceu')
else:
  print('empate')

```

```python
#h

perguntas = ['telefonou pra vitima?', 'esteve no local do crime?', 'mora perto da vitima?', 'devia para vitima?', 'ja trabalhou com a vitima?']

nivel_suspeito = 0
respostas = []

for x in perguntas:
  resposta = input(x)
  respostas.append(resposta)
  
for x in respostas:
  if(x.lower() == 'sim'):
    nivel_suspeito +=1

if (nivel_suspeito == 2):
  print ('suspeito')
elif (nivel_suspeito > 2 and nivel_suspeito < 4):
  print('cumplice')
elif (nivel_suspeito == 5):
  print('assassino')
else:
  print('inocente')

print(respostas)
```
